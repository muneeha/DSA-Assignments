## 4.Fill in all the missing values present in all the columns in the dataset



#### if missing values are more than 10 % of the values then better dropping rather than filling

```python
#Since the colum age is a Numerical column fill the colum with its Median()
data['Age']=data['Age'].fillna(data['Age'].median())
```

```python
data
```

```python
#Filling Categorical Colums with its Mode
data['Embarked']=data['Embarked'].fillna(data['Embarked'].mode()[0])
data['Cabin']=data['Cabin'].fillna(data['Cabin'].mode()[0])
```

```python
data
```

```python
#Dropping the cabin column because it has  more than 10 % null values
data.drop('Cabin',axis=1,inplace=True)
```

## 5. Checking and handling outliers in 3 columns in the dataset

#### Checking and handling outliers in the Fare Column

```python
plt.boxplot(data['Fare'])
plt.title('Boxplot Of Fare Columns')
```

```python
#From the boxplot we can understand thatoutlier present in the column Fare
```

```python
#handling outlier
Q1=np.percentile(data['Fare'],25,interpolation='midpoint')
Q2=np.percentile(data['Fare'],50,interpolation='midpoint')
Q3=np.percentile(data['Fare'],75,interpolation='midpoint')
```

```python
print(Q1)
print(Q2)
print(Q3)
```

```python
data['Fare'].median()
```

```python
IQR =Q3-Q1
```

```python
print(IQR)
```

```python
low_limit=Q1-1.5*IQR
up_limit=Q3+1.5*IQR
```

```python
print(low_limit)
print(up_limit)
```

```python
outlier=[]
for x in data['Fare']:
    if((x>up_limit)or(x<low_limit)):
        outlier.append(x)
```

```python
outlier
```

```python

#locating indexes of outliers and dropping it
index_out = data[(data['Fare']>up_limit)|(data['Fare']<low_limit)].index
```

```python
data.drop(index_out,inplace=True)
```

```python
data.head()
```
