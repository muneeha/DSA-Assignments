## Unsupervised learning models


### Kmeans Clustering

```python
from sklearn.cluster import KMeans
wcss=[]

for i in range(1,11):
  
    kmeans=KMeans(n_clusters = i, init = 'k-means++',random_state=42)
    kmeans.fit(scaled_data)
    wcss.append(kmeans.inertia_)
    # so wcss list will have wcss values with clusters 1 to 10 model
    centroids = kmeans.cluster_centers_
    
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

##### So from the elbow point we can understand that optimum no: of clusters is 3

```python
centroids
```

```python
kmeans =KMeans(n_clusters =3, init='k-means++',random_state=42)# code to create a k means model with 3 clusters
y_kmeans3=kmeans.fit_predict(scaled_data)
    
```

```python
y_kmeans3
```

```python
from sklearn.metrics import silhouette_score
sil_sc_3 = silhouette_score(scaled_data,y_kmeans3)
sil_sc_3
```

```python
kmeans =KMeans(n_clusters =2, init='k-means++',random_state=42)# code to create a k means model with 3 clusters
y_kmeans2=kmeans.fit_predict(scaled_data)
    
```

### DB SCAN--- is a density based Clustering

DBSCAN requires  and minPts parameters for clustering. The minPts parameter is easy to set. The minPts should be 4 for two-dimensional dataset. For multidimensional dataset, minPts should be 2 * number of dimensions. 

```python
from sklearn.neighbors import NearestNeighbors

# n_neighbors = 14 as kneighbors function returns distance of point to itself (i.e. first column will be zeros)
nbrs = NearestNeighbors(n_neighbors=14).fit(scaled_data)

# Find the k-neighbors of a point
neigh_dist, neigh_ind = nbrs.kneighbors(scaled_data)

# sort the neighbor distances (lengths to points) in ascending order
# axis = 0 represents sort along first axis i.e. sort along row
sort_neigh_dist = np.sort(neigh_dist, axis=0)
```

```python
nbrs.kneighbors(scaled_data)
```

```python
import matplotlib.pyplot as plt
k_dist = sort_neigh_dist[:,13]
plt.plot(k_dist)
plt.axhline(y=3.4, linewidth=1, linestyle='dashed', color='k')
plt.title("Plot of K-NN distance for Nearest neighbors")
plt.ylabel("k-NN distance")
plt.xlabel("Distance based sorted observations")
plt.show()
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=3.4, min_samples=26)
model=db.fit(scaled_data)
```

```python
# get cluster labels
labels=model.labels_
print(labels)
```

#-1 value represents noisy points could not assigned to any cluster

```python
# check unique clusters
set(model.labels_)
```

```python
silhouette_score(scaled_data,labels)
```

### Agglomerative Clustering
