**###5.Perform meaningful visualizations using the dataset. Bring  at least 3 visualizations**

```python
# A)Two Bar charts
## First showing showing Species of the flower,its Petal length and petal width
## Second showing showing Species of the flower,its Sepal length and Sepalwidth


plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.bar(data_flowers["Classification"],data_flowers["PL"],label='PL')
plt.bar(data_flowers["Classification"],data_flowers["PW"],label='PW')
plt.title("Plot showing Species of the flower,its Petal length and petal width")
plt.legend()

plt.figure(figsize=(8,8))
plt.subplot(2,1,1)
plt.bar(data_flowers["Classification"],data_flowers["SL"],label='SL')
plt.bar(data_flowers["Classification"],data_flowers["SW"],label='SW')
plt.title("Plot showing Species of the flower,its Sepal length and Sepal width")
plt.legend()

plt.tight_layout
```

```python
#B) Four Scatter plots each showing the relation of petal lenth,petal width,sepal length and sepal width to their specie type

#Plot1
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.title("Petal length of the flower with specie type")
sns.scatterplot(x=data_flowers.index,y=data_flowers["PL"], hue=data_flowers['Classification'],palette=("green","orange","blue"))


#Plot2
plt.figure(figsize=(12,8))
plt.subplot(2,2,2)
plt.title("Petal Width of the flower with specie type")
sns.scatterplot(x=data_flowers.index,y=data_flowers["PW"], hue=data_flowers['Classification'],palette=("red","cyan","purple"))

#Plot3
plt.figure(figsize=(12,8))
plt.subplot(2,2,3)
plt.title("Sepal width of the flower with specie type")
sns.scatterplot(x=data_flowers.index,y=data_flowers["SW"], hue=data_flowers['Classification'],palette=("green","orange","blue"))

#Plot4
plt.figure(figsize=(12,8))
plt.subplot(2,2,4)
plt.title("Sepal length of the flower with specie type")
sns.scatterplot(x=data_flowers.index,y=data_flowers["SL"], hue=data_flowers['Classification'],palette=("green","orange","blue"))

plt.tight_layout

```

```python
#C)Four box plots showing the different species with their respective petal length, petal width, sepal length and sepal width
#It gives the maximun values,minimum vales,mean.third quartile & first quartile


#Plot1
plt.figure(figsize=(8,8))
plt.subplot(2,2,1)
sns.boxplot(x=data_flowers["Classification"],y=data_flowers["PL"],hue=data_flowers["Classification"],palette=("red","green","blue"))
plt.title("Species of each flower with its Petal length" )

#Plot2
plt.figure(figsize=(8,8))
plt.subplot(2,2,2)
sns.boxplot(x=data_flowers["Classification"],y=data_flowers["PW"],hue=data_flowers["Classification"],palette=("yellow","gray","lightgreen"))
plt.title("Species of each flower with its Petal width" )

#Plot3
plt.figure(figsize=(8,8))
plt.subplot(2,2,3)
sns.boxplot(x=data_flowers["Classification"],y=data_flowers["SL"],hue=data_flowers["Classification"],palette=("red","green","blue"))
plt.title("Species of each flower with its Sepal length" )

#Plot4
plt.figure(figsize=(8,8))
plt.subplot(2,2,4)
sns.boxplot(x=data_flowers["Classification"],y=data_flowers["SW"],hue=data_flowers["Classification"],palette=("yellow","gray","lightgreen"))
plt.title("Species of each flower with its Sepal Width" )
```

```python
#D) Box plot of the Flowers data

plt.figure(figsize=(7,8))
sns.boxplot(data=data_flowers)
```

```python
#E) Two Scatter Plots


# First Scatter plot to check the correlation between Petal Length and Petal Width
#Here it is showing positive correlation

plt.figure(figsize=(8,6))
plt.subplot(2,1,1)
plt.scatter(x=data_flowers['PL'],y=data_flowers['PW'],color='green',marker='^')
plt.title('Correration between Petal length and Petal Width')
plt.xlabel=("PL")
plt.ylabel=("PW")
#Second Scatter plot to check the correlation between sepal length and sepal Width
#This doesnt show a clear positive correlation nor a clear negative correlation

plt.figure(figsize=(8,6))
plt.subplot(2,1,2)
plt.scatter(x=data_flowers['SL'],y=data_flowers['SW'],color='red',marker='^')
plt.title('Correration between Sepal Length and Sepal Width')
plt.xlabel=("SL")
plt.ylabel=("SW")


plt.tight_layout()
```
